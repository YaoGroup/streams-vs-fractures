```python
def count_files_in_folder(folder_path):
    try:
        # Get a set of the files actually present in the folder, excluding hidden files
        all_files = os.listdir(folder_path)
        files = [f for f in all_files if not f.startswith('.') and os.path.isfile(os.path.join(folder_path, f))]
        return len(files)
    except FileNotFoundError:
        print(f"The folder {folder_path} does not exist.")
        return 0

# Example usage
folder_path = "/data/data/v4_split_image/fracture_cleaned/"
num_files = count_files_in_folder(folder_path)
print(f"Number of files in folder and subfolders: {num_files}")
```

```python
def check_files_in_folder(folder_path, expected_files):
    try:
        # Get a set of the files actually present in the folder
        present_files = set(os.listdir(folder_path))
        present_files = {f for f in present_files if not f.startswith('.')}
        
        # Get sets of the expected files that are present and missing
        missing_files_in_folder = set(expected_files) - present_files
        unwanted_files_in_folder = present_files - set(expected_files)
        
        return missing_files_in_folder, unwanted_files_in_folder
    except FileNotFoundError:
        print(f"The folder {folder_path} does not exist.")
        return set(), set()

def check_files_in_folder(folder_path, expected_files):
    try:
        # Get a set of the files actually present in the folder
        present_files = set(os.listdir(folder_path))
        present_files = {f for f in present_files if not f.startswith('.')}
        
        # Get sets of the expected files that are present and missing
        missing_files_in_folder = set(expected_files) - present_files
        unwanted_files_in_folder = present_files - set(expected_files)
        
        return missing_files_in_folder, unwanted_files_in_folder
    except FileNotFoundError:
        print(f"The folder {folder_path} does not exist.")
        return set(), set()
```

```python
def create_tiff_from_png(png_path, source_tiff_path, output_tiff_path):
    # Open the source TIFF file to get metadata
    with rasterio.open(source_tiff_path) as src:
        metadata = src.meta.copy()
        transform = src.transform
        crs = src.crs

    # Open the PNG file
    png_image = Image.open(png_path).convert('L')  # Convert to grayscale if needed
    png_array = np.array(png_image)
    print(png_array.shape)

    # Update metadata for the new TIFF file
    metadata.update({
        'driver': 'GTiff',
        'height': png_array.shape[0],
        'width': png_array.shape[1],
        'count': 1,
        'dtype': png_array.dtype,
        'transform': transform,
        'crs': crs
    })

    # Write the new TIFF file
    with rasterio.open(output_tiff_path, 'w', **metadata) as dst:
        dst.write(png_array, 1)  # Write the image data to band 1
```
